```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
m = 1.3
b = 1
Xs = np.arange(0, 10, 0.1)
Ys = [(m * x) + b + np.random.normal() for x in Xs]
```

```python
plt.plot(Xs,Ys, '.')
plt.savefig('../fig/simple_lin_reg_GEN.jpg')
```

```python
plt.plot(Xs,Ys, '.')
plt.plot(Xs, m*Xs + b, '-')
plt.savefig('../fig/simple_lin_reg_w_line_GEN.jpg')
```

```python
Xs = np.arange(-1, 5, 0.1)
plt.plot(Xs, -(Xs-2)**2 + 2, '-')
plt.plot([2], [2], 'x', color='red', markersize=10)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.axvline(2, color='gray', linestyle='dashed')
plt.savefig('../fig/deriv_at_2_GEN.jpg')
```
